# Universal Swift UVOT Pipeline
**Edit only the `SOURCES` list below.** Everything else runs automatically.

For each source provide:
- `name` — used for output filenames
- `data_dir` — folder containing the raw `_mag.fits` files
- `ext_csv` — NED galactic extinction table (must have `Bandpass` and `Galactic Extinction` columns)
- `out_dir` — where to save the corrected CSV and plot

UV extinctions (W1, M2, W2) are derived automatically from E(B-V) = A_B − A_V using the Cardelli et al. (1989) law.

```python
# ============================================================
#  PIPELINE  —  do not edit below this line
# ============================================================

import os
import glob
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
from astropy.time import Time
from astropy.utils.exceptions import AstropyWarning

warnings.simplefilter('ignore', category=AstropyWarning)

# Cardelli et al. (1989) R_lambda = A_lambda / E(B-V) for Swift UVOT UV filters
R_UVOT = {'w1': 6.40, 'm2': 8.70, 'w2': 8.10}

SWIFT_MJD_REF = 51910.0007428


def load_extinction(ext_csv):
    """Read A_lambda for all 6 UVOT filters from a NED extinction table."""
    tbl = pd.read_csv(ext_csv)
    tbl['Bandpass'] = tbl['Bandpass'].str.strip()

    def get(name):
        row = tbl[tbl['Bandpass'] == name]
        if row.empty:
            raise ValueError(f"'{name}' not found in {ext_csv}")
        return float(row['Galactic Extinction'].iloc[0])

    A_V = get('Landolt V')
    A_B = get('Landolt B')
    A_U = get('Landolt U')
    ebv = A_B - A_V

    return {
        'v':  A_V,
        'b':  A_B,
        'u':  A_U,
        'w1': ebv * R_UVOT['w1'],
        'm2': ebv * R_UVOT['m2'],
        'w2': ebv * R_UVOT['w2'],
    }


def filter_from_filename(fname):
    f = fname.lower()
    if   'um2' in f: return 'm2'
    elif 'uw1' in f: return 'w1'
    elif 'uw2' in f: return 'w2'
    elif 'uuu' in f: return 'u'
    elif 'ubb' in f: return 'b'
    elif 'uvv' in f: return 'v'
    return None


def get_mjd(hdul, table):
    for hdu in hdul:
        hdr = hdu.header
        for key in ('MJD-OBS', 'MJD_OBS'):
            if key in hdr:
                return float(hdr[key])
        if 'DATE-OBS' in hdr:
            return Time(hdr['DATE-OBS'], format='isot').mjd
    for col in ('TSTART', 'MET', 'TIME'):
        if col in table.colnames:
            val = float(table[col][0])
            if val > 0:
                return SWIFT_MJD_REF + val / 86400.0
    return None


def extract_source(src):
    """Extract, correct, and save CSV for one source. Returns the DataFrame."""
    name     = src['name']
    data_dir = src['data_dir']
    out_dir  = src['out_dir']
    out_csv  = os.path.join(out_dir, f'{name}_UVOT_Corrected.csv')

    extinction = load_extinction(src['ext_csv'])

    print(f"\n{'='*55}")
    print(f"  {name}")
    print(f"{'='*55}")
    print("  Extinction (A_lambda in mag):")
    for f, a in extinction.items():
        print(f"    {f.upper():3s}: {a:.4f}")

    fits_files = glob.glob(os.path.join(data_dir, '*.fits'))
    print(f"\n  Found {len(fits_files)} FITS files.")

    rows, skipped, non_det = [], 0, 0
    for fpath in fits_files:
        filt = filter_from_filename(os.path.basename(fpath))
        if filt is None:
            skipped += 1
            continue
        try:
            with fits.open(fpath) as hdul:
                t   = Table.read(fpath, hdu=1)
                mjd = get_mjd(hdul, t)
                if mjd is None:
                    skipped += 1
                    continue
                mag = float(np.ma.filled(t['MAG'][0], 99.0))
                if mag >= 99.0:
                    non_det += 1
                    continue
                raw_flux = float(t['FLUX_AA'][0])
                raw_err  = float(t['FLUX_AA_ERR'][0])

            factor = 10 ** (extinction[filt] / 2.5)
            rows.append({
                'MJD':            round(mjd, 5),
                'Filter':         filt.upper(),
                'Intrinsic_Flux': raw_flux * factor,
                'Flux_Error':     raw_err  * factor,
            })
        except Exception as e:
            print(f"  [!] {os.path.basename(fpath)}: {e}")
            skipped += 1

    df = pd.DataFrame(rows).sort_values('MJD').reset_index(drop=True)
    os.makedirs(out_dir, exist_ok=True)
    df.to_csv(out_csv, index=False)

    print(f"\n  Detections saved : {len(df)}")
    print(f"  Non-detections   : {non_det}")
    print(f"  Skipped          : {skipped}")
    print(f"  Output           : {out_csv}")
    print(f"\n  Per-filter counts:")
    for filt, grp in df.groupby('Filter'):
        print(f"    {filt}: {len(grp)}  "
              f"[{grp['Intrinsic_Flux'].min():.3e} — {grp['Intrinsic_Flux'].max():.3e}]")

    return df


def plot_source(src, df):
    """Plot all available UVOT bands for one source."""
    name    = src['name']
    out_dir = src['out_dir']

    # ── Apply MJD window from config ─────────────────────────
    mjd_min = src.get('mjd_min')
    mjd_max = src.get('mjd_max')
    if mjd_min is not None:
        df = df[df['MJD'] >= mjd_min]
    if mjd_max is not None:
        df = df[df['MJD'] <= mjd_max]

    available = [f for f in FILTERS if f in df['Filter'].values]
    if not available:
        print(f"  No data to plot for {name} in the specified MJD range.")
        return

    fig, axes = plt.subplots(len(available), 1,
                              figsize=(12, 2.5 * len(available)),
                              sharex=True)
    if len(available) == 1:
        axes = [axes]

    color_map = dict(zip(FILTERS, COLORS))

    for ax, filt in zip(axes, available):
        subset = df[df['Filter'] == filt]
        ax.errorbar(x=subset['MJD'], y=subset['Intrinsic_Flux'],
                    yerr=subset['Flux_Error'],
                    fmt='o', color=color_map[filt], ecolor='gray',
                    capsize=2, markersize=4, linestyle='None',
                    label=f'Swift {filt}')
        ax.set_ylabel('Flux\n(erg/s/cm²/Å)', fontsize=9)
        ax.legend(loc='upper right', fontsize=9)
        ax.grid(True, linestyle='--', alpha=0.5)

    # Show the active MJD range in the x-label
    if mjd_min is not None or mjd_max is not None:
        lo = f"{mjd_min}" if mjd_min is not None else "start"
        hi = f"{mjd_max}" if mjd_max is not None else "end"
        axes[-1].set_xlabel(f'Time (MJD)   [{lo} – {hi}]', fontsize=11)
    else:
        axes[-1].set_xlabel('Time (MJD)', fontsize=11)

    title = name.replace('_', ' ')
    plt.suptitle(f'{title} — Swift UVOT Light Curves', fontsize=13, y=1.01)
    plt.tight_layout()

    out_jpg = os.path.join(out_dir, f"{name}_Multiband.jpg")
    plt.savefig(out_jpg, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"  Plot saved: {out_jpg}")


# ── Run all sources ──────────────────────────────────────────
for src in SOURCES:
    df = extract_source(src)
    plot_source(src, df)
```
